```python
import folium as fl
```

```python
centro_bahia = [-12.5, -41.5]
```

```python
mapa = fl.Map(location=centro_bahia, zoom_start = 6)
```

```python
cidades = {
    "Remanso": [-9.6231, -42.0805],
    "Feira de Santana": [-12.2669, -38.9668],
    "ibotirama": [-12.1852, -43.2215],
    "Barreiras": [-12.1466, -44.9961],
    "Irêce": [-11.3023, -41.8569],
    "Jacobina": [-11.1818, -40.5188],
    "Santo Antônio de jesus": [-12.9682, -39.2614],
    "Salvador": [-12.9714, -38.5114],
    "Ipiaú": [-14.1353, -39.7368],
    "Vitória da Conquista": [-14.8661, -40.8394],
    "Itabuna": [-14.7963, -39.2736]
}
```

```python
for cidade, coord in cidades.items():
   fl.Marker(
      location=coord,
      popup = cidade,
      tooltip = f"Ver {cidade}",
      icon=fl.Icon(color="blue", icon="info-sign")
      ).add_to(mapa)
```

```python
mapa.save("../src/assets/mapa_bahia_lafmed.html")
print(f"Mapa gerado com sucesso! ")
```
